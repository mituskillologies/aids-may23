#### Visualize the data

```python
sns.jointplot(x='Budget',y='Celebrities',hue='Class',data=df)
```

#### Build the model

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier()
```

```python
x = df[['Budget','Celebrities']]
y = df['Class']
```

```python
clf.fit(x, y)
```

```python
new1 = [[5600000, 2]]
clf.predict(new1)
```
