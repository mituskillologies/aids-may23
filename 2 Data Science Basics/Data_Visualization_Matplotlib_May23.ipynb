```python
import matplotlib.pyplot as plt
import numpy as np
```

### Line Plot

```python
x = np.array([4,13,18,32,44,49,58,61,74,88])
y = np.array([10,21,37,39,41,59,67,71,83,90])
```

```python
plt.plot(x, y)
```

```python
z = np.array([89,76,65,59,52,51,42,38,33,21])
```

```python
plt.plot(x, z)
```

```python
plt.figure(figsize=(16,9))
plt.title('My first plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.grid()
plt.xticks(range(0,101,10))
plt.yticks(range(0,101,10))
plt.xlim(0,100)
plt.ylim(0,100)
# marker:  .  o  *  >  <  v  ^  D  
plt.plot(x, y, marker='D', color='red', ls=':', lw=3, ms=10,
        mfc='green')
```

```python
plt.figure(figsize=(16,9))
plt.title('My first plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.grid()
plt.plot(x, y, color = 'red', label='2023')
plt.plot(x, z, color = 'blue', label='2022')
plt.legend()
```

```python
plt.figure(figsize=(16,9))
plt.subplot(2,1,1)
plt.plot(x, y, color = 'red')
plt.subplot(2,1,2)
plt.plot(x, z, color = 'blue')
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.plot(x, y, color = 'red')
plt.subplot(1,2,2)
plt.plot(x, z, color = 'blue')
plt.savefig('myplot.png')
```

### Scatter Plot

```python
plt.figure(figsize=(16,9))
plt.title('My first plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.grid()
plt.scatter(x, y, marker='o')
```

```python
python = np.array([67,82,46,41,62,78,85,40,73,63,41,68])
java = np.array([61,85,41,49,69,80,85,45,71,69,47,70])
```

```python
plt.grid()
plt.xlabel('Java Marks')
plt.ylabel('Python Marks')
plt.scatter(python, java)
```

### Bar plot

```python
studs = [78,45,122,167,69,90]
courses = ['C','C++','Java','Python','R','PHP']
```

```python
plt.grid()
plt.xlabel('Course Name')
plt.ylabel('Total Enrollment')
plt.bar(courses, studs, color='brown')
```

```python
plt.grid()
plt.ylabel('Course Name')
plt.xlabel('Total Enrollment')
plt.barh(courses, studs, color='green')
```

### Pie Chart

```python
plt.figure(figsize=(10,10))
plt.pie(studs, labels=courses, autopct='%2.2f%%');
```

## Univariate Analysis

### Histogram

```python
marks = np.array([67,81,64,62,45,58,52,89,67,63,58,60,62,68])
```

```python
plt.xlabel('Range of Marks')
plt.ylabel('Count')
plt.xlim(0,100)
plt.grid()
plt.xticks(range(0,101,10))
plt.hist(marks, bins=range(1,101,10));
```

### Box plot

```python
plt.boxplot(marks);
```
