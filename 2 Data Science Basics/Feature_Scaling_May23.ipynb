```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
df.columns
```

```python
plt.xlabel('Age')
plt.ylabel('Salary')
plt.scatter(df['Age'], df['EstimatedSalary'], c=df['Purchased'])
```

#### Normalization

```python
# import the class
from sklearn.preprocessing import MinMaxScaler
```

```python
# Create the object
scaler = MinMaxScaler()
```

```python
# Apply the scaling
x = df[['Age','EstimatedSalary']]
x_scaled = scaler.fit_transform(x)
```

```python
x_scaled[:,0];
```

```python
x_scaled[:,1];
```

```python
plt.xlabel('Age')
plt.ylabel('Salary')
plt.scatter(x_scaled[:,0], x_scaled[:,1], c=df['Purchased'])
```

```python
x_scaled[:,1].min(), x_scaled[:,1].max()
```

```python
x_scaled[:,0].min(), x_scaled[:,0].max()
```

#### Standardization

```python
# import the class
from sklearn.preprocessing import StandardScaler
# Create the object
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
x_scaled
```

```python
plt.xlabel('Age')
plt.ylabel('Salary')
plt.scatter(x_scaled[:,0], x_scaled[:,1], c=df['Purchased'])
```
